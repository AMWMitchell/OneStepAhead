# Data Visualisation - EDA

```python
import pandas as pd
import numpy as np
import altair as alt
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="ticks", palette="pastel")
```

```python
df = pd.read_csv('../../data/processed_data/cases_airports.csv')
```

```python
df
```

```python
ax = sns.lineplot(x="days", y="growth_rate", data=df)
```

```python
plt.figure(figsize=(28, 16))
sns.lineplot(x="days", y="growth_rate", hue='Country/Region', estimator=None, data=df)
```

```python
# Show the results of a linear regression within each dataset
sns.lmplot(x="days", y="growth_rate", col="Country/Region", hue="Country/Region", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```
