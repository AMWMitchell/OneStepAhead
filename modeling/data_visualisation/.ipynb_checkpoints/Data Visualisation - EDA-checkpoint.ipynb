# Data Visualisation - EDA

```python
import pandas as pd
import numpy as np
import altair as alt
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="ticks", palette="pastel")
```

```python
df = pd.read_csv('../../data/processed_data/cases_airports.csv')
```

```python
df
```

```python
set(df['Country/Region'])
```

```python
#country_list_depricated = ['Germany' , 'US', 'United Kingdom', 'South Africa', 'Cuba', 'Namibia']
```

```python
country_list = ['China','Japan','South Korea','UK','US','Italy','Iran','Germany','Algeria','Egypt','Burkina Faso','South Africa','Brazil']
```

```python
data_to_plot = df[df['Country/Region'].isin(country_list)]
```

```python
plt.figure(figsize=(28, 16))
plt.xscale('log')
plt.yscale('log')
ax = sns.lineplot(x="days", y="growth_rate", data=data_to_plot)
```

```python
plt.figure(figsize=(28, 16))
plt.xscale('log')
plt.yscale('log')
ax = sns.lineplot(x="cases_cum", y="cases", hue='Country/Region', data=data_to_plot)
```

```python
plt.figure(figsize=(28, 16))
plt.xscale('log')
plt.yscale('log')
sns.lineplot(x="days", y="growth_rate", hue='Country/Region', estimator=None, data=data_to_plot)
```

```python
plt.figure(figsize=(28, 16))
plt.xscale('log')
plt.yscale('log')
sns.lineplot(x="days_since_first_case", y="growth_rate", hue='Country/Region', estimator=None, data=data_to_plot)
```

```python
# Show the results of a linear regression within each dataset
sns.lmplot(x="days", y="growth_rate", col="Country/Region", hue="Country/Region", data=data_to_plot,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```
