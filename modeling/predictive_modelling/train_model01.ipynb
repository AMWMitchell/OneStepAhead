```python
import pandas as pd
import numpy as np
%matplotlib inline

from sklearn import preprocessing
from sklearn.metrics import mean_squared_log_error
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
import seaborn as sns
```

## get data ready

```python
df = pd.read_csv('../../data/processed_data/cases_airports.csv')
```

```python
df = df.drop(columns=['index', 'Lat', 'Long', 'date', 'cases_cum', 'growth_rate']) 
```

```python
df
```

### new feature - last days cases

```python
df['last_day_cases'] = df.groupby(['Country/Region'])['cases'].shift().fillna(0)
```

```python
df
```

```python
len(set(df['Country/Region']))
```

## Baseline model

### define metric measurement

```python
def rmsle(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))
```

### make naive forecast (cases on day to predict = cases day before)

```python
mean_error = []
for day in range(59, 66):
    train = df[df['days'] < day]
    val = df[df['days'] == day]

    p = val['last_day_cases'].values

    error = rmsle(val['cases'].values, p)
    print('day %d - Error %.5f' % (day, error))
    mean_error.append(error)
print('Mean Error = %.5f' % np.mean(mean_error))
```

## Data preparation for traing

### establish label encoder

```python
le = preprocessing.LabelEncoder()
```

```python
#ohe = preprocessing.OneHotEncoder(handle_unknown='ignore')
```

### fit transform le

```python
df['Country/Region'] = le.fit_transform(df['Country/Region'])
```

```python
len(le.classes_)
```

```python
df
```

## Train the model

### establish simple Random Forest Regressor

### train

```python
mean_error = []
y_preds = []
y_tests = []
X_tests = []

model = RandomForestRegressor(n_estimators=2500, n_jobs=-1, random_state=42)

for day in range(59, 66):
    train = df[df['days'] < day]
    test = df[df['days'] == day]
    
    X_train, X_test = train.drop(['cases'], axis=1), test.drop(['cases'], axis=1)
    y_train, y_test = train['cases'].values, test['cases'].values

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    
    error = rmsle(y_test, y_pred)
    print('day %d - Error %.5f' % (day, error))
    mean_error.append(error)
    y_preds.append(y_pred)
    y_tests.append(y_test)
    X_tests.append(X_test)
print('Mean Error = %.5f' % np.mean(mean_error))
```

```python
mean_error = []
y_preds = []
y_tests = []
X_tests = []

for day in range(59, 66):
    train = df[df['days'] < day]
    test = df[df['days'] == day]

    X_train, X_test = train.drop(['cases'], axis=1), test.drop(['cases'], axis=1)
    y_train, y_test = train['cases'].values, test['cases'].values

    model = RandomForestRegressor(n_estimators=2500, n_jobs=-1, random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    
    error = rmsle(y_test, y_pred)
    print('day %d - Error %.5f' % (day, error))
    mean_error.append(error)
    y_preds.append(y_pred)
    y_tests.append(y_test)
    X_tests.append(X_test)
print('Mean Error = %.5f' % np.mean(mean_error))
```
